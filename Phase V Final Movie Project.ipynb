## Duration
I want to examine the relationship between duration and average score of the movie because the scatterplot showed a linear relationship in the middle of the spread. However, the data is more spread out towards the ends, making the overall relationship non-linear.

```python
list_dur = np.sort(movie_data["duration"]) 
list_count = pd.unique(list_dur)
list_group = [movie_data.where(movie_data["duration"] == i) for i in list_count]
avg_list = [i.imdb_score.mean() for i in list_group]
plt.scatter(list_count, avg_list, alpha = 0.5, color = 'c')
plt.xlabel("Movie Duration")
plt.ylabel("Average IMDB Score")
plt.title("Movie Duration vs Average IMDB Score")
```

```python
dur_score_p = movie_data["duration"].corr(movie_data["imdb_score"], method = "pearson")
dur_score_s = movie_data["duration"].corr(movie_data["imdb_score"], method = "spearman")
dur_score_cov = movie_data["duration"].cov(movie_data["imdb_score"])
print("The Pearson Correlation between movie length and rating: {:.2f}".format(dur_score_p))
print("The Spearman Correlation between movie length and rating: {:.2f}".format(dur_score_s))
print("The Covariance between movie length and rating: {:.2f}".format(dur_score_cov))
```

The relationship between the movie duration and the score is 0.34, which is a moderately weak correlation. The covariance and correlations indicate that the duration has a positive relationship with the score. Below, I am performing linear regression on duration vs the average scores. 

```python
dur = pd.DataFrame(list_count, columns = ["duration"])
dur["avg_score"] = avg_list
feat_d =  dur[["duration"]]
target_d = dur["avg_score"]

x_train_d,x_test_d, y_train_d, y_test_d = train_test_split(feat_d, target_d, test_size = 0.2, random_state = 42)

duration_model = LinearRegression()
duration_model.fit(x_train_d, y_train_d)
predict = duration_model.predict(x_test_d)

print("Score: ", duration_model.score(x_test_d, y_test_d))

fig = plt.figure()
ax = fig.add_subplot(111)
orig = ax.scatter(list_count, avg_list, alpha = 0.5, color = 'c')
test = ax.scatter(x_test_d, y_test_d, color = 'r')
plt.plot(x_test_d, predict, 'k', linewidth = 4)
plt.legend([orig, test], ["original data", "testing data"])
plt.xlabel("Duration")
plt.ylabel("Average Score")
plt.title("Linear Regression on Duration vs Average Score")

```

The score of the model is pretty low, and the plotted line of predicttions does not correlate with the original data. I am going to remove outliers from the data to see if this changes the regression.

```python
# dur = pd.DataFrame(list_count, columns = ["duration"])
# dur["avg_score"] = avg_list

col = dur["duration"]
dur = dur[(dur["duration"] >= col.quantile(.15)) & (dur["duration"] <= col.quantile(.85))]

feat_d2 =  dur[["duration"]]
target_d2 = dur["avg_score"]

x_train_d2,x_test_d2, y_train_d2, y_test_d2 = train_test_split(feat_d2, target_d2, test_size = 0.2, random_state = 42)

duration_model2 = LinearRegression()
duration_model2.fit(x_train_d2, y_train_d2)
predict2 = duration_model2.predict(x_test_d2)

print("Score: ", duration_model2.score(x_test_d2, y_test_d2))

fig = plt.figure()
ax = fig.add_subplot(111)
a = ax.scatter(dur["duration"], dur["avg_score"], alpha = 0.5, color = 'c')
b = ax.scatter(x_test_d2, y_test_d2, color = 'r')
plt.plot(x_test_d2, predict2, 'k', linewidth = 4)
plt.legend([a,b], ["Original Data", "Testing Data"])
plt.xlabel("Duration")
plt.ylabel("Average Score")
plt.title("Linear Regression of Duration vs Average Score with Removed Outliers")




```

Linear regression is a good model for the data without outliers. The score is about 0.7, indicating that the model is doing moderately well for the data. A linear model makes sense for the data. As the movie length increases, the score should increase because people would not want to watch a movie that is too long or too short. However, when the movies become too long, less people would enjoy watching as well. This us explained by the data because around 140 minutes, the scores are more spread out and there is a less direct increase in score, as duration increases. This indicates that for longer movies, the score is less predictable, since the lower scores may be from people who believed the movies were too long and the higher scores may be from people who genuinely enjoy the content of the movie. On the other end, lower durations had more spread out scores in the first scatterplot. This makes sense because when movies are too short, people are less likely to enjoy the movie, so the lower scores will be from people who did not enjoy the legnth and higher scores will be from people who really enjoyed the content. Looking at this graph, it would appear that an optimal movie duration is about 140-150 minutes because this is right before the data begins to become very varied. 

Based on these findings, duration can function as a moderately good feature for predicting the score.

```python
genre_data = movie_data.copy()
genre_data['sing_genre'] = np.zeros((len(genre_data), 1))
```

```python
def is_genre(col_name, genre):
    '''
    Creates a new column to represent whether the row is of a specific genre. 
    
    Parameters: col_name is the name of the column. 
    Precondition: col_name is a string. 
    
    Parameters: genre is the name of the movie genre.
    Precondition: genre is a string with the first letter capitalized.
    '''
    genre_data[col_name] = np.zeros((len(genre_data), 1))
    for i in range(len(genre_data['genres'])): 
        if genre_data['genres'][i].find(genre) != -1: 
            genre_data[col_name][i] = 1
            genre_data['sing_genre'][i] = genre
        else:
            genre_data[col_name][i] = 0
    genre_data[col_name] = genre_data[col_name].astype(int)
```

```python
## How to isolate genres 
is_genre('is_adventure', 'Adventure')
is_genre('is_drama', 'Drama')
is_genre('is_comedy', 'Comedy')
is_genre('is_thriller', 'Thriller')
is_genre('is_horror', 'Horror')
is_genre('is_romance', 'Romance')

for i in range(len(genre_data['sing_genre'])): 
    if (genre_data['sing_genre'][i] == 0.0):
        genre_data['sing_genre'][i] = 'Other'
        
genre_data.head()
```

# Pre-registration 1: KMeans Clustering for Genres 

In this section I am using KMeans clustering to try creating clusters based on the number of users voted and the IMDb score. KMeans clustering takes in an input and uses centroids to group data based on similar means. 

```python
numerical_data = genre_data[['users_voted', 'imdb_score']]
data_with_genre = genre_data[['users_voted', 'imdb_score', 'sing_genre']]

kmeans = KMeans(n_clusters = 7)
kmeans.fit(numerical_data)
y_kmeans = kmeans.predict(numerical_data)
```

The code used to graph the plots are from Cornell Data Science Intro to Machine Learning Course from Lecture 9.

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.figure(figsize = (20,10))
_, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

# original graph
for i in data_with_genre['sing_genre'].unique():
    data = data_with_genre[data_with_genre['sing_genre'] == i]
    ax1.plot(data[['users_voted']],data[['imdb_score']],'o')
ax1.set_title("Movies by Genre")
ax1.set_xlabel('Number of Users Voted')
ax1.set_ylabel('Score')

# kmeans graph
for i in range(len(data_with_genre['sing_genre'].unique())):
    data = numerical_data.iloc[np.where(labels==i)]
    ax2.plot(data['users_voted'],data['imdb_score'],'o')
    lines = plt.plot(centroids[i,0],centroids[i,1],'kx')
ax2.set_title("Movie Genres by K-Means Clustering")
ax2.set_xlabel('Number of Users Voted')
ax2.set_ylabel('Score')

plt.show()
```

In the model, I used the 7 genres as clusters, to see if the genres could be classified from the data. The graphs clearly change after K-Means clustering and there are 7 centroids. However, I want to see what the best value of k is (the best number of clusters). 

I got the elbow function code from https://realpython.com/k-means-clustering-python/#evaluating-clustering-performance-using-advanced-technique, which helped me determine the best k-value from an elbow graph and graph it. 
    

```python
kmeans_kwargs = {"init": "random","n_init": 10, "max_iter": 300,"random_state": 42}
sse = []
for k in range(1, 11):
   kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
   kmeans.fit(numerical_data)
   sse.append(kmeans.inertia_)
```

SSE is the squared estimate of errors which is the sum of the squares of the residuals. In this loop, the total squares of the points' distances away from the cluster is determined and appended to a list. Smaller SSE's are better because it means that the data is closer to the centroids. 

```python
plt.figure(figsize = (7,5))
plt.plot(range(1, 11), sse, color = 'c')
plt.xticks(range(1, 11))
plt.title('Elbow Graph of Clusters')
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

The elbow is the point when the graph stops dropping rapidly. This is because eventually, adding more clusters will not change the error in the graph. The elbow point of the graph appears to be at 3 clusters. This indicates that the best number of clusters is 3 clusters for the model. Below I make a model with 3 cluster.

```python
kmeans_opt = KMeans(n_clusters = 3)
kmeans_opt.fit(numerical_data)
y_kmeans_opt = kmeans.predict(numerical_data)
```

```python
labels = kmeans_opt.labels_
centroids = kmeans_opt.cluster_centers_

plt.figure(figsize = (20,10))
_, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

# original graph
for i in data_with_genre['sing_genre'].unique():
    data = data_with_genre[data_with_genre['sing_genre'] == i]
    ax1.plot(data[['users_voted']],data[['imdb_score']],'o')
ax1.set_title("Movies by Genre")
ax1.set_xlabel('Number of Users Voted')
ax1.set_ylabel('Score')

# kmeans graph
for i in range(3):
    data = numerical_data.iloc[np.where(labels==i)]
    ax2.plot(data['users_voted'],data['imdb_score'],'o')
    lines = plt.plot(centroids[i,0],centroids[i,1],'kx')
ax2.set_title("Genre by K-Means Clustering")
ax2.set_xlabel('Number of Users Voted')
ax2.set_ylabel('Score')

plt.show()
```

# Significance Testing

## Permutation Tests

I want to analyze the variables that all seemed to have low correlation and see if these were just by chance or if the data was significant. Using permutation tests will help me do this because it rearranges the data by shuffling the outputs. I will be comparing the permutations with the null hypothesis. 

To perform the permutation tests, I am using Professor Wilken's code from the lecture 3/31 code. 

```python
def permute(input_array):
    permuted = input_array.copy()
    np.random.shuffle(permuted)
    return permuted
```

## Duration

```python
model = LinearRegression().fit(movie_data[["duration"]], movie_data["imdb_score"])
```

```python
model = LinearRegression().fit(movie_data[["duration"]], movie_data["imdb_score"])
print("Coefficient: ", model.coef_[0])
```

```python
steeper_slopes = 0
n_permutations = 10000
permuted_slopes = np.zeros(n_permutations)

for i in range(n_permutations):
    fake_model = LinearRegression().fit(movie_data[["duration"]], permute(movie_data["imdb_score"]))
    
    permuted_slopes[i] = fake_model.coef_[0]
    
    if np.abs(fake_model.coef_[0]) > np.abs(model.coef_):
        steeper_slopes += 1
    
print(steeper_slopes / n_permutations)

```

```python
binsizes, _, _ = plt.hist(permuted_slopes, bins=20, color = 'c', alpha = 0.7)
plt.vlines(model.coef_[0], ymin=0, ymax=np.max(binsizes), color='red')
plt.title("Permuted Slopes")
plt.xlabel("coefficient")
plt.ylabel("count")
plt.show()
```

Using permutation tests on the duration, the slopes of the permuted data and the original model are very different. The model coefficient is outside of the permuted slopes distribution, indicating that the slope is significant. The p-value I calculated was 0. Since the p-value is less thatn 0.05, I reject the null hypothesis. It is unlikely for the data to ahve occurred under the null hypothesis, so there is a lower probability of the duration data to occur by chance.

This is interesting because of my earlier data. The permutation test indicates that I accept the alternative hypothesis - that the duration is increasing the score.

## Bootstrap Resampling Method

```python
sample_data = movie_data[["users_voted", "imdb_score"]]
```

```python
n_bootstrap = 10000
bootstrap_slopes = np.zeros(n_bootstrap)

for i in range(n_bootstrap):
    resampled_data = sample_data.sample(n=len(sample_data), replace=True)
    fake_model = LinearRegression().fit(resampled_data[["users_voted"]], resampled_data["imdb_score"])
    
    bootstrap_slopes[i] = fake_model.coef_[0]
    
    
```

While originally, the relationship between rank and score seemed extremely obvious, since rank is derived from score, the scatter plot and the models show me that the score does not constantly decrease as ranking decreases.

__Linear regression models with more features did not increase the accuracy of the models.__ 

Originally, I predicted that the users voted would indicate a more popular movie with an either very high score or a very low score. While the correlation between number of users voted and score have a relatively weak correlation, it is not high enough to make a good model. I wanted to look into the number of genres because a higher number of genres would appeal to more people and gain more popularity, or a certain genre would turn people away from liking the movie. However, genre had a very low correlation because most of the genres had average score. Duration showed to have a stronger model score, as well as an optimal duration of around 140-150 minutes. I also wanted to use the year in case certain years had higher scores.

Since the correlations between number of users voted, number of reviews, duration, and number of genres had low correlation with the score individually, I tried to perform linear regression with multiple features. I decided not to use both number of users voted and number of reviews together because they are very related, which I showed in my data. In the model with number of users voted, duration, and number of genres, there was a score of around 18.4%. This is a pretty weak score for the model. However, the second model had a similar score of about 17.8%. Most of the data is around average in the data with removed outliers, so it is hard to predict the score based of different number of reviews. 

Overall, I could not create a good model with the features of users voted, duration, and number of genres. There still seems to be very little relation between users voted, duration and number of genres. 

__Permutation testing showed whether the data was significant or not.__ 

I performed permutation tests on duration, number of genres, number of votes. It appears that the number of votes, genre, and duration was significant since the model coefficient was outside of the distributions. This makes sense because duration and number of votes had higher correlations, even if they were not strong. 

I expected the number of genres to be not have a correlation with the score because of how low a correlation there is. While I thought that increasing the number of genres would increase the score, there did not seem to be a correlation that indicated this relationship. The permutation test indicated that the data is significant and the model was not plausible under the null hypothesis, so the low correlation is likely not due to chance. 

The number of votes made sense because the scatterplot and barplot showed and positive relationship between the score and the number of people who voted on the movie. The duration hypothesis test did not make sense to me because of when I broke up the data into 3 parts. There did not seem to be a strong correlation between the duration and the score, except for longer durations. The linear regression models were also fairly weak for the duration models.

__Bootstrap resampling shows how my data is reproducible.__ 

The boostrap method indicated that the coefficient of my original linear regression model for the number of users voted and the movie scores is within the confidence interval produced by the resampling. The relationship between the number of users voted and the movie scores is able to be reproduced, as well as the linear regression model. My conclusion for this is that people will not watch lower rated movies because they do not want to watch movies. While there are strong opinions for really low-scored movies and really-highly scored movies, the highly negative opinions become clouded by people not watching the movies. Towards the middle of the distribution, where the movie scores are not really high and are not really low, people are still willing to watch and they my give a score really high or really low after watching the movie. There is an increasing trend with the increasing number of voters, indicating that more people will vote on positive movies.
